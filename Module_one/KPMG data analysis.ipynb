## transactions_df data analysis

```python
# check null values
transactions_df.isnull().sum()
```

```python
transactions_df.columns.values
```

```python
# find out whether the 197 null values came from the same data points
null_columns = transactions_df[['brand', 'product_line',
       'product_class', 'product_size','standard_cost',
       'product_first_sold_date']]
```

```python
null_columns_df = null_columns[null_columns.isnull().any(axis=1)]
```

```python
# get the data that doesn't data for online orders
online_order_df = transactions_df.loc[transactions_df['online_order'].isnull()]
```

```python
overlap_null = pd.merge(online_order_df.reset_index(), null_columns_df.reset_index(), how='inner', on=["index"])
```

```python
# null values that overlap between online orders and the null group
overlap_null
```

```python
# all null values in the dataset
transactions_df[transactions_df.isnull().any(axis=1)].head()
```

```python
transactions_df.head()
```

```python
transactions_df.describe(include="all")
```

```python
transactions_df.info()
```

```python
transactions_missing_values1 = transactions_df.iloc[null_columns[null_columns.isnull().any(axis=1)].index]
transactions_missing_values2 = online_order_df
transactions_missing_values_overlap = overlap_null.drop(columns=['index'])
```

```python
transactions_missing_values_overlap
```

```python
transactions_missing_values1.to_csv("transactions_worksheet/transactions_missing_values1.csv", index=False)
transactions_missing_values2.to_csv("transactions_worksheet/transactions_missing_values2.csv", index=False)
transactions_missing_values_overlap.to_csv("transactions_worksheet/overlap_missing_values.csv", index=False)
```
